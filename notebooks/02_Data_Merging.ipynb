```python
import pandas as pd

df_first = pd.read_csv('dataset/dataset_01.1/dataset01.1.csv')
df_second = pd.read_csv('dataset/dataset_01.2/dataset01.2.csv')

# List the datasets in the order you want them stacked
frames = [df_first, df_second]

# axis=0 tells pandas to stack them vertically (rows)
df_total = pd.concat(frames, axis=0, ignore_index=True)

# Save the new, larger dataset
df_total.to_csv('dataset/main_dataset/data.csv', index=False)

print(f"Dataset 1 rows: {len(df_first)}")
print(f"Dataset 2 rows: {len(df_second)}")
print(f"Total rows now: {len(df_total)}")
```
